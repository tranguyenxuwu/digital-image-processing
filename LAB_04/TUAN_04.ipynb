```python
#1 chọn đối tượng trong ảnh
import numpy as np
import imageio.v2 as iio
import matplotlib.pylab as plt
data = iio.imread('fruit.jpg')
bmg = data[800:1200, 570:980] #cắt ảnh từ hàng 800 đến 1200 và cột 570 đến 980
print (data.shape)

#iio.imsave('fruit_crop.jpg', bmg)
plt.imshow(bmg)
plt.show()
```

```python
#1.2 tịnh tiến đơn
import scipy.ndimage as nd

data = iio.imread("fruit.jpg", mode="F") # tuong duong as_gray=True 
bdata = nd.shift(data, (100, 25)) # tịnh tiến 100 hàng xuống dưới và 25 cột sang phải

plt.imshow (bdata)
plt.show()
```

```python
#1.3 thay doi kich thuoc anh
data = iio.imread('fruit.jpg')
print(data.shape)
bdata = nd.zoom(data, 2)
print(bdata.shape)
data2 = nd.zoom (data, (2,2,1))
print(data2.shape)
data3 = nd.zoom(data, (0.5, 8.5, 1))

plt.imshow (data3)
plt.show()
```

```python
#1.4 xoay anh
data = iio.imread('fruit.jpg')
print(data.shape)

dl = nd.rotate(data, 20) # xoay 20 do
plt.imshow(dl)
plt.show()

d2 = nd.rotate(data, 20, reshape=False) # xoay 20 do khong thay doi kich thuoc
plt.imshow(d2)
plt.show()
```

```python
#1.5. dilation va erosion
data = iio.imread('world_cup.jpg', mode="F")
print(data.shape)
a = nd.binary_dilation(data)
plt.imshow (a)
plt.show()

b = nd.binary_erosion(data, iterations=3)
plt.imshow(b)
plt.show()
```

```python
#1.6 coordinate mapping
data = iio.imread('world_cup.jpg', mode="F")
print(data.shape)

V, H = data.shape
M = np.indices((V, H))

d = 5
q = 2 * d * np.random.ranf(M.shape) - d
mp = M + q.astype(int)
d1 = nd.map_coordinates(data, mp)
plt.imshow(d1)
plt.show()
```

```python
#1.7 Generic Transformation

def GeoFun (outcoord) :
    a = 10 * np.cos(outcoord[0]/10.0) + outcoord[0]
    b= 10 * np.cos(outcoord[1]/10.0) + outcoord[1]
    return a, b

data = iio.imread("world_cup.jpg", mode="F")
d1 = nd.geometric_transform(data,GeoFun)
plt.imshow(d1)
plt.show()
```
