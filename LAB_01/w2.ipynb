```python
import numpy as np
import matplotlib.pyplot as plt
import imageio as io
import colorsys
import scipy.ndimage as sn
```

```python
a = io.imread('exercise/balloons_noisy.png')
k = np.ones((5, 5)) / 25
b = sn.convolve(a, k).astype(np.uint8)
io.imwrite('exercise/balloons_blurred.png', b)
# Display the original and blurred images
print(b)
plt.imshow(a)
```
